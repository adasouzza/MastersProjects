# GBFS

```python
# Sessão de bibliotecas
import copy
import time
```

## Função de imprimir Quebra-Cabeças

```python
def printNode(node):
    # imprimindo quebra-cabeças (utilizando index do vetor para organizar)
    print(node[0],node[1],node[2])
    print(node[3],node[4],node[5])
    print(node[6],node[7],node[8])
    global valorNode
    # imprimindo valor dos index, profundidade e caminho sendo percorrido
    print('Peca:', valorNode)
    print('Profundidade:', len(node[9:]))
    print('Caminho:', node[9:])
    print('------')
    valorNode += 1
```

## Função de Check de Estado Final

```python
def checkFinal(node):
    if node[:9]==estadoFinal:
        printNode(node)
        return True
    if node[:9] not in visitados:
        printNode(node)
        estadoAtual.append(node)
        visitados.append(node[:9])
    return False
```

## Função Heurística de Objetivo

```python
# Utilizando o estado objetivo como heuristica para a busca
def objetivoHeuristica(node):
    distancia = 0
    for current, target in enumerate(node):
        linha = int(current/3)
        coluna = current%3
        alvoLinha = int(target/3)
        alvoColuna = target%3
        distancia += abs(linha-alvoLinha) + abs(coluna-alvoColuna)
    return distancia
```

## Busca Gulosa

```python
if __name__ == '__main__':
    estadoInicial = [1,5,4, 3,7,2, 6,8,0]
    estadoFinal = [0,1,2, 3,4,5, 6,7,8]

    objetivo = False
    valorNode = 0
    visitados = []
    estadoAtual = []
    estadoAtual.append(estadoInicial)
    visitados.append(estadoInicial)
    printNode(estadoInicial)
    t0 = time.time()

    # enquanto o estado atual não for o objetivo realiza busca gulosa
    while (not objetivo and not len(estadoAtual)==0):
        fList = []
        for node in estadoAtual:
            h = objetivoHeuristica(node[:9])
            g = len(node[9:])
            f = g+h
            fList.append(f)
        estado = estadoAtual.pop(fList.index(min(fList)))
        indicePeca = estado.index(0)
    # MOVIMENTO PARA CIMA
        if indicePeca!=0 and indicePeca!=1 and indicePeca!=2:
            moverCima = copy.deepcopy(estado)
            moverCima[indicePeca] = moverCima[indicePeca-3]
            moverCima[indicePeca-3] = 0
            moverCima.append('CIMA')
            objetivo = checkFinal(moverCima)
    # MOVIMENTO PARA ESQUERDA
        if indicePeca!=0 and indicePeca!=3 and indicePeca!=6 and objetivo==False:
            moverEsquerda = copy.deepcopy(estado)
            moverEsquerda[indicePeca] = moverEsquerda[indicePeca-1]
            moverEsquerda[indicePeca-1] = 0
            moverEsquerda.append('ESQUERDA')
            objetivo = checkFinal(moverEsquerda)
    # MOVIMENTO PARA BAIXO
        if indicePeca!=6 and indicePeca!=7 and indicePeca!=8 and objetivo==False:
            moverBaixo = copy.deepcopy(estado)
            moverBaixo[indicePeca] = moverBaixo[indicePeca+3]
            moverBaixo[indicePeca+3] = 0
            moverBaixo.append('BAIXO')
            objetivo = checkFinal(moverBaixo)
    # MOVIMENTO PARA DIREITA
        if indicePeca!=2 and indicePeca!=5 and indicePeca!=8 and objetivo==False:
            moverDireita = copy.deepcopy(estado)
            moverDireita[indicePeca] = moverDireita[indicePeca+1]
            moverDireita[indicePeca+1] = 0
            moverDireita.append('DIREITA')
            objetivo = checkFinal(moverDireita)
    # imprimindo tempo decorrido        
    t1 = time.time()
    print('Tempo decorrido:', t1-t0)
    print('------')
```
